#### Q2. Write a Python function to estimate the population mean using a sample mean and standard deviation.

```python
def estimate_population_mean(sample_mean, sample_std, sample_size):
    import math
    
    standard_error = sample_std / math.sqrt(sample_size)
    z_critical = 1.96  # for 95% confidence level
    
    lower_bound = sample_mean - z_critical * standard_error
    upper_bound = sample_mean + z_critical * standard_error
    
    population_mean = (lower_bound + upper_bound) / 2
    
    return population_mean
```

```python
estimate_population_mean(3.4,2,39)
```

#### Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom parameter set to 10.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
```

```python
df = 10
```

```python
x = np.linspace(-3,3,100)
```

```python
y = t.pdf(x,df)
```

```python
plt.plot(x,y)
plt.title('Students t-distribution with df=10')
plt.xlabel('X-values')
plt.ylabel('PDF')
plt.show()
```

#### Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.

```python
import numpy as np
from scipy.stats import ttest_ind
```

```python
sample1 = np.random.normal(loc=10,scale=2,size=100)
```

```python
sample2 = np.random.normal(loc=12,scale=2,size=100)
```

```python
t_stat,p_value = ttest_ind(sample1,sample2)
```

```python
t_stat
```

```python
p_value
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("reject H0")
else:
    print("Fail to reject")
```

#### Q12: What is t-statistic? State the formula for t-statistic.

The t-statistic is a statistical measure used to determine the probability of whether a sample mean differs significantly from a hypothesized population mean.

t = (x̄ - μ) / (s / √n)

* x̄ is the sample mean
* μ is the hypothesized population mean
* s is the sample standard deviation
* n is the sample size
