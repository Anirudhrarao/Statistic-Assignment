#### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

In simpler terms, the central limit theorem tells us that if we take many random samples from a population and calculate the means of each sample, the distribution of these means will be approximately normal,

1. Allow for hypothesis tesing.
2. Provides a foundation for confidence interval estimation.
3. Enable statistical inference.
4. Allow for modeling and simulation.

#### Q10: State the assumptions of the Central Limit Theorem.

1. The random variable must be independent of each other.
2. The random variable must be same probability.
3. The variance of random variable must be finite.
4. The sample size must be large to ensure that mean approaches a normal distribution.
